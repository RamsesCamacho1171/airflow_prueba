Importamos las variables requeridas

```python
import pandas as pd
import math

data = {
    "link":"https://datosabiertos.impi.gob.mx/IMPICifras/Proteccion01.csv",
    "identificador":"Proteccion"
}

df=pd.read_csv(data['link'], encoding ="ISO-8859-1", delimiter=',', header = None)
```

```python
df.head()
```

Método para eliminar columnas en pandas
    axis = 1 Elimina columnas
    axis = 0 Elimina filas
    inplace = True Guarda los cambios en el dataFrame

```python
df.drop([0, 1, 2], axis=0, inplace = True)
df.head()
```

Comprobamos si la fila 4 columna 0 tiene o no tiene datos

```python
data = pd.isna(df[0][4])
print(data)
comprobar = "Es un valor NaN" if data == True else "No es un valor NaN"
print(comprobar)
```

```python
x = 0
longitud = len(df.columns)
while x < longitud:
    if pd.isna(df[x][3]) == True:
        df[x][3] = df[x - 1][3]
    x = x + 1
df.head()
```

Renombramos las columnas

```python
df.rename(columns={
    0: "Año",
    1: "Patentes",
    2: "Diseños industriales",
    3: "Modelos de utilidad",
    4: "Total"
}, inplace = True)
df.head()
```

Añadimos el identificador de la Base de datos

```python
df["Relacion"] = data['identificador']
df.head()
```

Reordenams las columnas

```python
df = df[['Año', 'Relacion', 'Patentes', 'Diseños industriales', 'Modelos de utilidad', 'Total']]
df.head()
```
